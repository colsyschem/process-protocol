```python
import glob
import os, time
import itertools
import numpy as np
import seaborn as sns
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.cm as cm
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
from cycler import cycler
from datetime import datetime, timedelta

from scipy.stats import norm
from scipy.optimize import curve_fit #we could import more, but this is what we need
from scipy.stats import linregress

from misc_process import normalize_max

%matplotlib inline
```

```python
'''
Get paths to working directory. This is the name of folder
in which are placed data.
'''

DIR = '20230620__working_title' 

# get the path to the project's main directory (e.g., Project051)
BASE_DIR_PATH = os.path.dirname(os.path.abspath(os.getcwd()))

# ------------------------- data folder -----------------------

# get the path to the folder containing raw data
DATA_DIR_PATH = os.path.join(BASE_DIR_PATH, 'data', '')
EXP_DIR_PATH = os.path.join(BASE_DIR_PATH, 'data', DIR,'')
UV_DIR_PATH = os.path.join(BASE_DIR_PATH, 'data', DIR,'UV-Vis-NIR','')

# ------------------------- output folder -----------------------

# # get the path to the folder with generated figures
PLOTS_DIR_PATH = os.path.join(BASE_DIR_PATH, 'results', 'plots', '')
```

```python
'''
Import example UV-Vis data from DIR'
'''

#================ import data =========================#

# import file using read_csv function in pandas
df = pd.read_csv(UV_DIR_PATH+'Example_UV_Vis_Rods.csv',
                 index_col= 'Wavelength (nm)')

# normalize all columns using home made function in misc_process.py
df_norm = normalize_max(df)
```

```python
'''
Make a simple figure containing normalized data 
'''
#================ plot data ===========================#

fig = plt.figure(figsize=(4,2), dpi=600) # generate a figure
ax = plt.subplot(1,1,1) #(nrow, ncolumn, index) - make a subplot, there is only one

ax.plot(df_norm)                           # plot data
ax.set_xlim([400,1100])                    # set wavelength limit
ax.set_ylim([0,1.1])                       # set absorbance limit
ax.set_xlabel(r'Wavelength (nm)')          # set axis label
ax.set_ylabel(r'Absorbance ')              # set axis label
ax.legend(df_norm.columns, 
          fontsize='small',
          frameon=False)

# #================ plot stylisation ====================#

ax.spines['right'].set_visible(False)      # remove spines
ax.spines['top'].set_visible(False)

plt.tight_layout()  
fig.savefig(PLOTS_DIR_PATH+f'UVVis_all__{DIR}.pdf')
```

```python
'''
Make a simple four-suplots figure
'''
#================ plot data ===========================#

fig = plt.figure(figsize=(6,4), dpi=600)
gs = fig.add_gridspec(2, 2) #nrows, #ncolumns
ax1 = fig.add_subplot(gs[0, 0]) # upper left subplot
ax2 = fig.add_subplot(gs[0, 1]) # upper rigth subplot
ax3 = fig.add_subplot(gs[1, 0]) # lower left subplot
ax4 = fig.add_subplot(gs[1, 1]) # lower right subplot

axes = [ax1, ax2, ax3, ax4] # list of all axes in plot
colors = ['firebrick', 'palegreen','skyblue', 'royalblue']
labels = ['sphere', 'cube', 'decahedra', 'rod']

for column, ax, color,label in zip(df_norm.columns, axes,colors, labels):
    ax.plot(df_norm[column], color=color)
    ax.set_xlim([400,1100])
    ax.set_ylim([0,1.1])
    ax.set_xlabel(r'Wavelength (nm)')
    ax.set_ylabel(r'Absorbance ')

# #================ plot stylisation ====================#

    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)

plt.tight_layout()  
fig.savefig(PLOTS_DIR_PATH+f'UVVIS_2by2__{DIR}.pdf')
```
